# Generate the data

# Data Dictionary

- **id**: Unique identifier for the export
- **dispcntry**: Country code of the dispatching country
- **destcntry**: Country code of the destination country
- **cmdtycode**: Commodity code of the goods
- **totvalue**: Total value of the goods
- **netmass**: Net mass of the goods in Kg
- **declarant_ref**: Unique identifier for the declarant
- **gdsdesc**: Description of the goods
- **clrncdate**: Clearance date
- **latitude**: Latitude of the destination country
- **longitude**: Longitude of the destination country

```python
# Define the column names based on the data dictionary
columns = ['id', 'dispcntry', 'destcntry', 'cmdtycode', 'totvalue', 'netmass', 'declarant_ref', 'gdsdesc', 'clrncdate',
            'latitude', 'longitude']

# Generate random data for 30 rows
data = []
for _ in range(300):
    # Randomly choose a destination country
    destcntry = random.choice(list(country_lat_long.keys()))
    # Get the corresponding latitude and longitude
    lat, long = country_lat_long[destcntry]

    # Randomly choose a cmmodity code
    cmdtycode = random.choice(list(commodity_codes.keys()))
    gdsdesc = commodity_codes[cmdtycode]

    row = {
        # Randomly generated pallet level identification number (EAN-8)
        'id': fake.ean(length=8),
        # Value of goods in Â£
        'totvalue': round(random.uniform(1000, 50000), 2),
        # Net mass of goods in kg
        'netmass': round(random.uniform(0.5, 1000), 2),
        'cmdtycode': cmdtycode,
        'gdsdesc': gdsdesc,
        'clrncdate': fake.date_this_year(),  # Random clearance date within this year
        'dispcntry': "GB",
        'destcntry': destcntry,
        # Random unique declarant reference (UUID)
        'declarant_ref': fake.uuid4(),
        'latitude': lat,
        'longitude': long
    }
    data.append(row)

# Create the DataFrame
df = pd.DataFrame(data, columns=columns)

# Ensure that the date column is in datetime format
df['clrncdate'] = pd.to_datetime(df['clrncdate'])

# Display the DataFrame
display(df.head())
```

```python
# Sort commodity codes by description
sorted_commodity_codes = sorted(commodity_codes.items(), key=lambda x: x[1])

# Sort destination countries by country code
sorted_dest_countries = sorted(df['destcntry'].unique())

# Create the Dash app
app = dash.Dash(__name__)

# App layout
app.layout = html.Div([
    html.H1("Commodity Dashboard"),

    # Dropdown to filter by commodity type with "All" option, sorted alphabetically
    html.Label("Select Commodity Type:"),
    dcc.Dropdown(
        id='commodity-dropdown',
        options=[{'label': 'All', 'value': 'All'}] + [{'label': desc,
                                                       'value': code} for code, desc in sorted_commodity_codes],
        value=['All'],  # Default to "All"
        multi=True,   # Allow multiple selections
        placeholder="Select commodity types"
    ),

    # Dropdown to filter by destination country with "All" option, sorted alphabetically
    html.Label("Select Destination Country:"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': 'All', 'value': 'All'}] + [{'label': code,
                                                       'value': code} for code in sorted_dest_countries],
        value=['All'],  # Default to "All"
        multi=True,   # Allow multiple selections
        placeholder="Select destination countries"
    ),

    # Date range filter
    html.Label("Select Date Range:"),
    dcc.DatePickerRange(
        id='date-picker',
        min_date_allowed=df['clrncdate'].min(),
        max_date_allowed=df['clrncdate'].max(),
        start_date=df['clrncdate'].min(),
        end_date=df['clrncdate'].max(),
        display_format='YYYY-MM-DD'
    ),
    # Map container
    dcc.Graph(
        id='proportional-symbol-map',
        style={'height': '600px'}  # Adjust the height of the map here
    ),

    # Display the filtered table
    dash_table.DataTable(id='filtered-table',
                         columns=[{'name': col, 'id': col}
                                  for col in df.columns],
                         page_size=10)  # Set number of rows per page

])

# Callback to handle "All" option logic based on selection order


@app.callback(
    Output('commodity-dropdown', 'value'),
    Output('country-dropdown', 'value'),
    Input('commodity-dropdown', 'value'),
    Input('country-dropdown', 'value'),
)
def update_dropdown_values(commodity_values, country_values):
    ctx = dash.callback_context  # To track which input triggered the callback
    triggered_input = ctx.triggered[0]['prop_id'].split('.')[0]

    # Determine if "All" was selected before other options or vice versa
    def process_values(values):
        # if all options have been removed, select 'All'
        if len(values) == 0:
            return ['All']

        if 'All' in values and len(values) > 1:
            # Check if "All" was selected first or last by its position in the list
            if values.index('All') == 0:
                # Remove "All" if other values were added
                return [v for v in values if v != 'All']
            else:
                # Keep only "All" if it was selected after other values
                return ['All']
        return values  # No change if "All" wasn't selected with other values

    # Process commodity dropdown
    if triggered_input == 'commodity-dropdown':
        commodity_values = process_values(commodity_values)

    # Process country dropdown
    if triggered_input == 'country-dropdown':
        country_values = process_values(country_values)

    return commodity_values, country_values

# Callback to update the table and the map based on filters


@app.callback(
    [Output('filtered-table', 'data'),
     Output('proportional-symbol-map', 'figure')],
    [Input('commodity-dropdown', 'value'),
     Input('country-dropdown', 'value'),
     Input('date-picker', 'start_date'),
     Input('date-picker', 'end_date')]
)
def update_table_and_map(selected_commodities, selected_countries, start_date, end_date):
    # Filter the DataFrame
    filtered_df = df.copy()

    # Handle commodity type filter
    if 'All' not in selected_commodities:
        filtered_df = filtered_df[filtered_df['cmdtycode'].isin(
            selected_commodities)]

    # Handle country filter
    if 'All' not in selected_countries:
        filtered_df = filtered_df[filtered_df['destcntry'].isin(
            selected_countries)]

    # Filter based on date range
    filtered_df = filtered_df[
        (filtered_df['clrncdate'] >= pd.to_datetime(start_date)) &
        (filtered_df['clrncdate'] <= pd.to_datetime(end_date))
    ]

    # Return the filtered data for the table
    table_data = filtered_df.to_dict('records')

    # Create the proportional symbol map
    if len(filtered_df) > 0:
        # Aggregate the data to get the total totvalue per destcntry
        map_data = filtered_df.groupby(['destcntry', 'latitude', 'longitude'], as_index=False).agg(
            {'totvalue': 'sum'}
        )

        # Dynamically determine a reasonable size_max based on the range of the totvalue
        size_max_value = 10

        # Create the map using Plotly Express
        fig = px.scatter_geo(
            map_data,
            lat='latitude',
            lon='longitude',
            size='totvalue',
            hover_name='destcntry',
            # Disable latitude and longitude in hover popup
            hover_data={'latitude': False, 'longitude': False},
            size_max=size_max_value,
            title='Total Value of Goods by Destination Country'
        )

        # Update layout for country names and borders
        fig.update_geos(
            showcountries=True,        # Show country borders
            countrycolor="Black",      # Border color for countries
            showcoastlines=True,       # Show coastlines
            coastlinecolor="Gray",     # Color of coastlines
            showland=True,             # Show land color
            landcolor="Gray",    # Color of the land areas
            projection_type="natural earth"  # Set projection type to Natural Earth
        )
        # Limit the map to a world view
        fig.update_layout(geo=dict(scope='world'))
    else:
        fig = px.scatter_geo(
            title='Proportional Symbol Map - No Data to Display')
        fig.update_geos(
            showcountries=True,
            countrycolor="Black",
            showcoastlines=True,
            coastlinecolor="Gray",
            showland=True,
            landcolor="LightGreen",
            projection_type="natural earth"
        )
        fig.update_layout(geo=dict(scope='world'))  # World view with no data

    return table_data, fig


# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
